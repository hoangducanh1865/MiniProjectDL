```python
import pandas as pd
from IPython.display import display

train_path = "../data/train.pkl"
test_path = "../data/test.pkl"

train_df = pd.read_pickle(train_path)
test_df = pd.read_pickle(test_path)

def preview_data(name, data, n=5):
    print(f"\n{name} type: {type(data)}")

    if isinstance(data, pd.DataFrame):
        print(f"{name} shape: {data.shape}")
        display(data.head(n))
        return

    if isinstance(data, dict):
        keys = list(data.keys())
        print(f"{name} keys ({len(keys)}): {keys}")

        shown = False
        for key, value in data.items():
            if isinstance(value, pd.DataFrame):
                print(f"{name}['{key}'] shape: {value.shape}")
                display(value.head(n))
                shown = True
                break
            if hasattr(value, "shape") and not shown:
                print(f"{name}['{key}'] shape: {value.shape}")
                try:
                    display(pd.DataFrame(value).head(n))
                except Exception:
                    print(f"First {n} values of {name}['{key}']:", value[:n])
                shown = True
                break

        if not shown and keys:
            first_key = keys[0]
            first_value = data[first_key]
            print(f"Preview {name}['{first_key}']:", first_value)
        return

    if hasattr(data, "shape"):
        print(f"{name} shape: {data.shape}")

    try:
        display(pd.DataFrame(data).head(n))
    except Exception:
        print(f"Cannot render head() for {name}. Raw preview:", data)

preview_data("Train", train_df)
preview_data("Test", test_df)
```

```python
import pandas as pd
from IPython.display import display

# 1. Định nghĩa đường dẫn tới các file pkl
train_path = "../data/train.pkl"
test_path = "../data/test.pkl"

# 2. Đọc file pickle
try:
    train_data = pd.read_pickle(train_path)
    test_data = pd.read_pickle(test_path)
    print("Đọc file thành công!")
except FileNotFoundError as e:
    print(f"Lỗi: Không tìm thấy file. Vui lòng kiểm tra lại đường dẫn! \nChi tiết: {e}")
    train_data, test_data = None, None

# 3. Hàm xem trước dữ liệu linh hoạt (Hỗ trợ DataFrame, Dict, Ndarray...)
def preview_data(name, data, n=5):
    if data is None:
        return
        
    print(f"\n" + "="*40)
    print(f" KHẢO SÁT DỮ LIỆU: {name.upper()}")
    print(f"="*40)
    print(f"- Kiểu dữ liệu gốc: {type(data)}")

    # TH1: Dữ liệu là một DataFrame thông thường
    if isinstance(data, pd.DataFrame):
        print(f"- Kích thước (Shape): {data.shape}")
        print(f"- Hiển thị {n} dòng đầu tiên:")
        display(data.head(n))
        return

    # TH2: Dữ liệu là Dictionary (thường dùng để gom nhiều tập dữ liệu)
    if isinstance(data, dict):
        keys = list(data.keys())
        print(f"- Dữ liệu dạng Dictionary gồm {len(keys)} keys: {keys}")
        
        for key, value in data.items():
            print(f"\n--> Kiểm tra Key: '{key}'")
            print(f"    + Kiểu dữ liệu: {type(value)}")
            if hasattr(value, "shape"):
                print(f"    + Kích thước (Shape): {value.shape}")
            
            # Hiển thị nếu là DataFrame hoặc mảng có thể đưa về DataFrame
            if isinstance(value, pd.DataFrame):
                display(value.head(n))
            elif hasattr(value, "__len__") and not isinstance(value, str):
                try:
                    display(pd.DataFrame(value).head(n))
                except Exception:
                    print(f"    + {n} giá trị đầu tiên: {list(value)[:n]}")
        return

    # TH3: Dữ liệu dạng mảng (Numpy array, List, Series...) nhưng không phải dataframe/dict
    if hasattr(data, "shape"):
        print(f"- Kích thước (Shape): {data.shape}")
    
    try:
        print(f"- Chuyển đổi hiển thị dưới dạng DataFrame ({n} dòng đầu):")
        display(pd.DataFrame(data).head(n))
    except Exception:
        print(f"- Không thể hiển thị dạng bảng. Bản in thô dữ liệu:")
        print(data[:n] if hasattr(data, "__getitem__") else data)

# 4. Thực thi hiển thị dữ liệu
preview_data("Train", train_data)
preview_data("Test", test_data)
```
